```python
# Imports
import pandas as pd
import os
from sqlalchemy import create_engine
```

## CSV to DataFrame

```python
# import health_csv 
health14_df = pd.read_csv('../Desktop/Health_Ins_2014.csv', encoding='latin-1')
health15_df = pd.read_csv('../Desktop/Health_Ins_2015.csv')
health16_df = pd.read_csv('../Desktop/Health_Ins_2016.csv')
health17_df = pd.read_csv('../Desktop/Health_Ins_2017.csv')
```

```python
health15_df.head()
```

## Transform

```python
# Health Columns wanted for analysis
health_standard14_df = pd.DataFrame(health14_df)
health_standard14 = health_standard14_df[ ['State','County','Metal Level','Issuer Name', 'Plan Type'] ]

health_standard15_df = pd.DataFrame(health15_df)
health_standard15 = health_standard15_df[ ['State Code','County Name','Metal Level','Issuer Name', 'Plan Type'] ]

health_standard16_df = pd.DataFrame(health16_df)
health_standard16 = health_standard16_df[ ['State Code','County Name','Metal Level','Issuer Name', 'Plan Type'] ]

health_standard17_df = pd.DataFrame(health17_df)
health_standard17 = health_standard17_df[ ['State Code','County Name','Metal Level','Issuer Name', 'Plan Type'] ]
```

### Create Database Connection

```python
connection_string = "postgres:postgres@localhost:5432/Project 2"
engine = create_engine(f'postgresql://{connection_string}')
```

###  Load DataFrames into database

```python
health_standard14.to_sql(name='HealthShop14', con=engine, if_exists='append', index=True)
health_standard15.to_sql(name='HealthShop15', con=engine, if_exists='append', index=True)
health_standard16.to_sql(name='HealthShop16', con=engine, if_exists='append', index=True)
health_standard17.to_sql(name='HealthShop17', con=engine, if_exists='append', index=True)
```
